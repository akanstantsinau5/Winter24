## (b) 

\
\begin{aligned}
& E\left[(\Delta W)^{2}\right]=\Delta t \Rightarrow u^{2} p_{u}+d^{2} p_{d}=\Delta t \Rightarrow \beta_{u}^{2} p_{u}+\beta_{d}^{2} p_{d}=1
\end{aligned}


\
\begin{aligned}
1. \Rightarrow p_{d}=1-p_{u}
\end{aligned}
\
\begin{aligned}
2. \Rightarrow \beta_{u} p_{u}=\beta_{d}(1-p_{u}) \Rightarrow p_{u}=\frac{\beta_{d}}{\beta_{u}+\beta_{d}}
\end{aligned}
\
\begin{aligned}
3. \Rightarrow \beta_{u}^{2} \cdot \frac{\beta_{d}}{\beta_{u}+\beta_{d}}+\beta_{d}^{2} \cdot \frac{\beta_{u}}{\beta_{u}+\beta_{d}}=1 \Rightarrow \beta_{d}=\frac{1}{\beta_{u}}
\end{aligned}

\
\begin{aligned}
Hence,\
p_{u}=\frac{1}{\beta_{u}^{2}+1}, \quad p_{d}=\frac{\beta_{u}^{2}}{\beta_{u}^{2}+1}
\end{aligned}

\
\begin{aligned}\\
& \text{(1)}\ \beta_{u}=1, \quad \beta_{d}=1, \quad p_{u}=\frac{1}{2}, \quad p_{d}=\frac{1}{2}\\
& \text{(2)}\ \beta_{u}=\sqrt{2}, \quad \beta_{d}=\frac{\sqrt{2}}{2}, \quad p_{u}=\frac{1}{3}, \quad p_{d}=\frac{2}{3}\\
& \text{(3)}\ \beta_{u}=2, \quad \beta_{d}=\frac{1}{2}, \quad p_{u}=\frac{1}{5}, \quad p_{d}=\frac{4}{5}
\end{aligned}
