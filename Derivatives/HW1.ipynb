### Problem 2:
Option Strategy Values and Sensitivities
plot the value, delta, and gamma as the spot price St is varied between 75 and 135 for the
following payoffs (all assumed long):
In each instance, describe whether the value or sensitivity is positive or negative (and
over what ranges). If it changes sign (or trends toward zero or some other asymptote), at
roughly what spot level does that occur? Is it monotonically increasing or decreasing or
does it display a maximum (or minimum)?

You are welcome to use numerically differentiated values (with sufficiently small pertur-
bations) for the sensitivities though closed-form expressions are – of course – preferred.
For reference, the BSM values for ATM calls and puts with St = K = 100 using these
conventions are C = 8.73949 and P = 6.79856.
You may find it helpful to refer to the LIFFE document distributed by Prof. Dahlquist.

```python
# Define Black_Scholes model as a function for convenience
def black_scholes(start, end, K, T, r, y, sigma, option_type):
    # Hande option type input
    if option_type.lower() == 'call':
        fi = 1
    elif option_type.lower() == 'put':
        fi = -1
    else:
        raise ValueError('Invalid input. Put or Strike')
    # Initiate empty list for plotting
    bs_price = []

    # Iterate through the values from start to end to prices
    for price in range(start, end):
        S = price
        # Calculate d1 and d2
        d1 = (np.log(S/K)+(r-y)*T)/(sigma*T)
        d2 = (sigma*T)/2
        # Calculate z's
        z_plus = d1 + d2
        z_minus = d1 - d2
        # Calculate N(d1) and N(d2)
        n_d1 = norm.cdf(fi*z_plus)
        n_d2 = norm.cdf(fi*z_minus)
        # Calculate deltas
        delta = fi*np.exp(-y*T)*n_d1
        # Calculate gamma
        gamma = (norm.pdf(d1)*np.exp(-y*T))/(S*sigma*np.sqrt(T))
        # Calculate the price
        bs = fi*((S*np.exp(-y*T)*n_d1)-(K*np.exp(-r*T)*n_d2))
        # Append values to the list
        bs_price.append({'Spot Price': S, 'Option Price': bs, 'Delta': delta, 'Gamma': gamma})
    
    # Create a dataframe for plotting
    df = pd.DataFrame(bs_price)

    return(df)
```

##### (a) call with K = 100

```python
call_100 = black_scholes(75, 136, 100, 1, .04, .02, .2, 'call')
fig, ax = plt.subplots(1, 3, figsize=(15, 4))  
ax[0].plot(call_100['Spot Price'], call_100['Option Price'], color='blue')
ax[0].set_xlabel('Spot Price')
ax[0].set_ylabel('Option Price')
ax[0].set_title(f'Option Price for strike K=100')
ax[1].plot(call_100['Spot Price'], call_100['Delta'],  color='red')
ax[1].set_xlabel('Spot Price')
ax[1].set_ylabel('Delta')
ax[1].set_title(f'Delta for strike K=100')
ax[2].plot(call_100['Spot Price'], call_100['Gamma'], color='green')
ax[2].set_xlabel('Spot Price')
ax[2].set_ylabel('Gamma')
ax[2].set_title(f'Gamma for strike K=100')
plt.tight_layout()
plt.show()
```

- option price: positive and monotonically increasing; reverses toward 0 and is less than 1 from \$78
- delta: positive and monotonically increasing; reverses toward 0 and is less than 0.1 from \$75; trends toward 1 and is larger than 0.9 from \$127
- gamma: positive; it displays a maximum roughly at \$94

##### (c) straddle with K = 100

To calculate our payoffs we also need to consider premiums that we pay (Call+Put of K=100). The payoff will be price versus. 

```python
# Create straddle DataFrame
straddle = pd.DataFrame()
# Set the spot price
straddle['Spot Price'] = call_100['Spot Price']
# Calculate Straddle Price
straddle['Straddle Price'] = call_100['Option Price'] + put_100['Option Price']
# Calculate Delta
straddle['Delta'] = call_100['Delta'] + put_100['Delta']
# Calculate Gamma
straddle['Gamma'] = call_100['Gamma'] + put_100['Gamma']
fig, ax = plt.subplots(1, 3, figsize=(15, 4))  
ax[0].plot(straddle['Spot Price'], straddle['Straddle Price'], color='blue')
ax[0].set_xlabel('Spot Price')
ax[0].set_ylabel('Straddle Price')
ax[0].set_title(f'Straddle Price for strike K=100')
ax[1].plot(straddle['Spot Price'], straddle['Delta'], color='red')
ax[1].set_xlabel('Spot Price')
ax[1].set_ylabel('Delta')
ax[1].set_title(f'Straddle Delta for strike K=100')
ax[2].plot(straddle['Spot Price'], straddle['Gamma'], color='green')
ax[2].set_xlabel('Spot Price')
ax[2].set_ylabel('Gamma')
ax[2].set_title(f'Straddle Gamma for strike K=100')
plt.tight_layout()
plt.show()
```

- option price: positive; it displays a minimum roughly at \$96 which is about the k=100 discounted by interest rate and dividend rate
- delta: changes sign from negative to positive around at \$96; has asymptote of -1 and 1
- gamma: positive; it displays a maximum roughly at \$94; same as double gamma of call or put with k = 100

##### d) 90-110 call spread
